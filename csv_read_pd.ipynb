```python
x = [2,3,4,55,6,4,3,2,7]

var = pd.Series(x, index=["a","b","c","d","c","d","e","f","e"], dtype = "float",name = "Python")

print(var)
print(type(var))
print(var[4])
```

# Write CSV

```python
import pandas as pd

dis = {
    'a' : [1,2,3,4,5],
    'b' : [3,5,2,6,10], # creating data using dictionary
    'c' : [6,7,8,9,0] 
}
d1 = pd.DataFrame(dis)

print(d1)

d1.to_csv("demo.csv") # using function .to_csv("file_name.csv")
# file saved at same directory.

                 
```

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'City': ['NY', 'LA', 'NY', 'LA', 'NY', 'LA'],
    'Sales': [100, 200, 150, 300, 250, 180],
    'Profit': [30, 70, 50, 90, 60, 40]
})

print(df)
```

```python
df.groupby("City").mean(numeric_only = True)
```

```python
df['Sales'] = pd.to_numeric(df['Sales'], errors='coerce')
df.groupby('City').mean(numeric_only = True)
```

```python
df.groupby('City').agg({
    'Sales': 'mean',
    'Profit': 'first'
})
```

```python
df.groupby('City')[['Sales']].mean()
```
