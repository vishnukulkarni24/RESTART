- Reverse a given number.
- Example:

- Input: 1234
- Output: 4321

```python
num = 1234
rev = 0
while num:
    rev = rev * 10 + num % 10
    num = num//10
print(rev)
```

Count the number of digits in an integer.
- Example:

Input: 789
Output: 3

```python
num = 789
length = 0
while num:
    length +=1
    num = num // 10
print(length)
```
