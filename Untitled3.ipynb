```python
import pandas as pd
import random
import numpy as np
from datetime import datetime , timedelta

df = pd.read_csv("C:\\Users\\Vishnu\\Desktop\\RESTART\\Heart_Data.csv", dtype= str)
df
```

```python
def random_date(start, end):
    delta = end - start
    random_days = random.randrange(delta.days)
    return start + timedelta(days = random_days)

formats = ["%d:%m:%y", "%m:%d:%y", "%y:%m:%d", "%d-%m-%Y", "%m/%d/%Y"]

start_date = datetime(2000, 1, 1)
end_date = datetime(2025, 12, 31)

random_dates = []
for _ in range(len(df)):
    date = random_date(start_date, end_date)
    fmt = random.choice(formats)
    random_dates.append(date.strftime(fmt))


df["Date"] = random_dates

df.to_csv("C:\\Users\\Vishnu\\Desktop\\RESTART\\Heart_Data.csv", index = False)

print(df.head)
```
